# Analyzing Data from the data frames

## Working with dates

```python
covid_df.date
```

```python
# Notice that the dtype is object. This means that pandas don't know that this column contains dates.
# So, we'll change this column dtype to datetime:

covid_df['date'] = pd.to_datetime(covid_df.date)
covid_df.date
```

You can see that it now has the datatype datetime64. We can now extract different parts of the data into separate columns, using the DatetimeIndex class.

```python
covid_df['year'] = pd.DatetimeIndex(covid_df.date).year
covid_df['month'] = pd.DatetimeIndex(covid_df.date).month
covid_df['day'] = pd.DatetimeIndex(covid_df.date).day
covid_df['weekday'] = pd.DatetimeIndex(covid_df.date).weekday
```

```python
covid_df
```

```python
covid_df_may = covid_df[covid_df.month == 5]
covid_df_may
```

```python
covid_df_may_metrics = covid_df_may[["new_cases", "new_deaths", "new_tests"]]
covid_df_may_metrics
```

```python
covid_may_totals = covid_df_may_metrics.sum()
covid_may_totals
```

```python
type(covid_may_totals)
```

```python
# The operations above can also be combined into a single statement

covid_df[covid_df.month == 5][["new_cases", "new_deaths", "new_tests"]].sum()
```

## Grouping and aggregation

```python
covid_month = covid_df.groupby('month')
covid_month
```

We have to aggregate after grouping. Let's aggregate using sum()

```python
covid_month_df = covid_df.groupby('month')[['new_cases', 'new_deaths', 'new_tests']].sum()
covid_month_df
```

```python
covid_month_mean_df = covid_df.groupby('weekday')[['new_cases', 'new_deaths', 'new_tests']].mean()
covid_month_mean_df
```

```python
covid_df['total_cases'] = covid_df.new_cases.cumsum()
covid_df['total_deaths'] = covid_df.new_deaths.cumsum()
covid_df['total_tests'] = covid_df.new_tests.cumsum() + 935310
```

```python
covid_df
```

## Merging data from multiple sources

```python
from urllib.request import urlretrieve
urlretrieve("https://hub.jovian.ml/wp-content/uploads/2020/09/locations.csv", "location.csv")
```

```python
location_df = pd.read_csv("location.csv")
location_df
```

```python
location_df[location_df.location == "Italy"]
```

```python
covid_df['location'] = 'Italy'
covid_df
```

```python
merged_df = covid_df.merge(location_df, on = "location")
merged_df
```

```python
merged_df['cases_per_thousand'] = merged_df.total_cases * 1e3 / merged_df.population
merged_df['deaths_per_thousand'] = merged_df.total_deaths * 1e3 / merged_df.population
merged_df['tests_per_thousand'] = merged_df.total_tests * 1e3 / merged_df.population
```

```python
merged_df
```

## Writing data back to files

```python
result_df = merged_df[[
    'date',
    'new_cases',
    'total_cases',
    'new_deaths',
    'total_deaths',
    'new_tests',
    'total_tests',
    'cases_per_thousand',
    'deaths_per_thousand',
    'tests_per_thousand',
    'location',
    'continent'
]]
```

```python
result_df
```

```python
result_df.to_csv('results.csv', index = None)
```
